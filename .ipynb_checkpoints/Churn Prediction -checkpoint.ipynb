```python
customerID=pd.get_dummies(churn_data['customerID'],prefix='customerID')
gender=pd.get_dummies(churn_data['gender'],prefix='gender')
Partner=pd.get_dummies(churn_data['Partner'],prefix='Partner')
Dependents	=pd.get_dummies(churn_data['Dependents'],prefix='Dependents')
PhoneService=pd.get_dummies(churn_data['PhoneService'],prefix='PhoneService')
MultipleLines=pd.get_dummies(churn_data['MultipleLines'],prefix='MultipleLines')
InternetService=pd.get_dummies(churn_data['InternetService'],prefix='InternetService')
OnlineSecurity=pd.get_dummies(churn_data['OnlineSecurity'],prefix='OnlineSecurity')
OnlineBackup=pd.get_dummies(churn_data['OnlineBackup'],prefix='OnlineBackup')
DeviceProtection=pd.get_dummies(churn_data['DeviceProtection'],prefix='DeviceProtection')
TechSupport=pd.get_dummies(churn_data['TechSupport'],prefix='TechSupport')
StreamingTV=pd.get_dummies(churn_data['StreamingTV'],prefix='StreamingTV')
StreamingMovies	=pd.get_dummies(churn_data['StreamingMovies'],prefix='StreamingMovies')
Contract=pd.get_dummies(churn_data['Contract'],prefix='Contract	')
PaperlessBilling=pd.get_dummies(churn_data['PaperlessBilling'],prefix='PaperlessBilling')
PaymentMethod=pd.get_dummies(churn_data['PaymentMethod'],prefix='PaymentMethod')
TotalCharges=pd.get_dummies(churn_data['TotalCharges'],prefix='TotalCharges')






```

```python
df1=pd.concat([customerID,gender,Partner,Dependents,PhoneService,MultipleLines,InternetService,OnlineSecurity,OnlineBackup,DeviceProtection,TechSupport,StreamingTV,Contract,PaperlessBilling,PaymentMethod,TotalCharges,churn_data.select_dtypes(['int64']),churn_data.select_dtypes(['float64'])],axis=1)
```

```python
churn_data.select_dtypes(['object'])
```

```python
df1.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(df1,y)
```

# SGDClassifier 

```python
from sklearn.linear_model import SGDClassifier
clf= SGDClassifier(random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_train,clf.predict(x_train))
```

```python
print(classification_report(y_train,clf.predict(x_train)))
```

```python
accuracy_score(y_test,clf.predict(x_test))
```

```python
print(classification_report(y_test,clf.predict(x_test)))
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred =cross_val_predict(clf,x_train,y_train,cv=3)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,y_train_pred)
```

# Decision Tree
